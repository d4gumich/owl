# Dowloading PDF Files and Extractiong text content.

```python
import pandas as pd
import json
import os
import requests
import pdfplumber
import uuid
```

#   Load report data

```python
# Load the NDJSON file into a DataFrame
report = pd.read_json('/content/drive/MyDrive/D4G Student Consulting Team/D4G Projects and Products/Project managment/PP-1_ReliefWeb_Bot_MVP/analysis_pages_2025-02-06.ndjson', lines=True)
report
```

# Dowload the PDF files and save them into a folder called PDFs

```python
# Define the directory for saving PDFs
pdf_directory = 'pdfs'

# Create the pdf_directory if it doesn't exist
if not os.path.exists(pdf_directory):
    os.makedirs(pdf_directory)
    print(f"Directory {pdf_directory} created")

# itinerate thru dataframe to get the files.
for index, row in report.iterrows():
    file_url = row['file']
    file_id = row['id']

    # Construct the filename
    filename = f"{file_id}.pdf"
    filepath = os.path.join(pdf_directory, filename)

    try:
        # Download the file
        response = requests.get(file_url, stream=True)
        response.raise_for_status()  # Raise an exception for bad status codes

        # Save the file
        with open(filepath, 'wb') as file:
            for chunk in response.iter_content(chunk_size=8192):
                file.write(chunk)
        print(f"Downloaded {filename} successfully.")

    except requests.exceptions.HTTPError as e:
        print(f"HTTP Error downloading {filename}: {e}")
    except requests.exceptions.RequestException as e:
        print(f"Request Error downloading {filename}: {e}")
    except Exception as e:
        print(f"An unexpected error occurred while downloading {filename}: {e}")


```

# Retrive text in chunks and create a dataframe

```python
def chunk_text(text, chunk_size=10560):
    """Divides text into chunks of a specified size."""
    return [text[i:i+chunk_size] for i in range(0, len(text), chunk_size)]

# Path to the directory containing PDFs
pdf_directory = 'pdfs'

# Prepare a list to hold all the data
data = []

# Process each PDF file in the directory
for filename in os.listdir(pdf_directory):
    if filename.endswith('.pdf'):
        # Remove the extension and any prefix from the filename for the ID
        file_id = filename.split('.')[0]
        file_id = file_id.split('_')[-1]  # Assuming a naming convention that includes a prefix

        filepath = os.path.join(pdf_directory, filename)
        try:
            with pdfplumber.open(filepath) as pdf:
                for page_number, page in enumerate(pdf.pages, start=1):
                    text = page.extract_text()
                    if text:  # Ensure there is text on the page
                        chunks = chunk_text(text)
                        for chunk_index, chunk in enumerate(chunks, start=1):
                            data.append({
                                'id': file_id,
                                'page_label': page_number,
                                'chunk_index': chunk_index,
                                'document': chunk,
                                'uuid': str(uuid.uuid4())
                            })
        except Exception as e:
            print(f"Failed to process {filename}: {e}")

```

```python
reports_pages= pd.DataFrame(data)
```

```python
def df_to_ndjson(input_df, output_ndjson_path):
    # Convert the entire DataFrame to a list of dictionaries at once
    records = input_df.to_dict('records')

    # Open the NDJSON file for writing
    with open(output_ndjson_path, 'w') as ndjson_file:
        # Iterate over the list of dictionaries and write each as a JSON object
        for record in records:
            json.dump(record, ndjson_file)
            ndjson_file.write('\n')  # Add a newline after each JSON object
```
