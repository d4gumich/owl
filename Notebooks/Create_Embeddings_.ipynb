# Create Embeddings for Search

```python
import pandas as pd
import json
import os
import requests
from sentence_transformers import SentenceTransformer
import uuid
```

```python
# Load the NDJSON file into a DataFrame
page_data = pd.read_json('/content/drive/MyDrive/D4G Student Consulting Team/D4G Projects and Products/Project managment/PP-1_ReliefWeb_Bot_MVP/page_data.ndjson', lines=True)
```

```python
reports = pd.read_json('/content/drive/MyDrive/D4G Student Consulting Team/D4G Projects and Products/Project managment/PP-1_ReliefWeb_Bot_MVP/analysis_pages_2025-02-06.ndjson', lines=True)
```

```python
page_data['document']=page_data['document'].str.replace('\n', '')
```

```python
combined_df = pd.merge(page_data, reports, on='id', how='left')
combined_df['year'] = pd.to_datetime(combined_df['date.created']).dt.year
```

```python
combined_df['combined_details'] = (combined_df['title'].astype(str) + combined_df['document'].astype(str) +" YEAR " + combined_df['year'].astype(str) +
                                    combined_df['origin'].astype(str) + " " + combined_df['primary_country_name'].astype(str) + " " +
                                    combined_df['source'].astype(str) + " " + combined_df['theme'].astype(str) + " " + combined_df['disaster'].astype(str))


```

```python
# Load the SBERT model
model = SentenceTransformer('all-MiniLM-L6-v2')

model

combined_df['embedding'] = combined_df['combined_details'].apply(lambda x: model.encode(x).tolist())

combined_df.head(2)

combined_df.to_csv('relief_web_embeddings.csv', index=False)
```
